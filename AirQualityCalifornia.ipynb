```python
import pandas as pd
from config import api_key
import requests
import sqlalchemy as sql
```

```python
air_quality_df=pd.read_csv('source_data/airquality_trends_city2000-2021_California.csv')
air_quality_df
```

```python
city='Bakersfield, California'
url=f'https://api.geoapify.com/v1/geocode/search?text={city}&format=json&apiKey={api_key}'
```

```python
response=requests.get(url)
data=response.json()
data
```

```python
lat=data['results'][0]['lat']
lon=data['results'][0]['lon']
```

```python
for idx, row in air_quality_df.head(150).iterrows(): 
    city=f'{row["CITY"]}, California'
    url=f'https://api.geoapify.com/v1/geocode/search?text={city}&format=json&apiKey={api_key}'
    response=requests.get(url)
    data=response.json()
    lat=data['results'][0]['lat']
    lon=data['results'][0]['lon']
    air_quality_df.loc[idx, 'Lat']=lat
    air_quality_df.loc[idx, 'Lon']=lon
```

```python
air_quality_df
```

```python
air_quality_df.to_csv('static/data/data.csv', index=False)
```

```python
engine=sql.create_engine('sqlite:///static/data/db.sqlite')
```

```python
air_quality_df.columns
```

```python
air_quality_df.to_sql('Pollutant', con=engine, index=False, if_exists='replace')
```

```python
result=engine.execute('select * from Pollutant').all()
type(result)
```
